# 층별 거래금액 평균 비교

```python
## 마이너스 표기된 층수 제외하고 1층 이상인 것들만 뽑기
floor2017 = apt2017['층'] >= 1
apt2017_floor = apt2017[floor2017]

floor2018 = apt2018['층'] >= 1
apt2018_floor = apt2018[floor2018]

floor2019 = apt2019['층'] >= 1
apt2019_floor = apt2019[floor2019]
```

```python
# 2017 새로운 열을 추가함
apt2017_floor['floor_group'] = apt2017_floor['층']

# floor_group은 문자열로 바꿔주기
apt2017_floor['floor_group'] = apt2017_floor['floor_group'].astype(str)

apt2017_floor.loc[(apt2017_floor['층'] > 0) & (apt2017_floor['층'] <= 5), 'floor_group'] = "1~5층"
apt2017_floor.loc[(apt2017_floor['층'] > 5) & (apt2017_floor['층'] <= 10), 'floor_group'] = "6~10층"
apt2017_floor.loc[(apt2017_floor['층'] > 10) & (apt2017_floor['층'] <= 15), 'floor_group'] = "11~15층"
apt2017_floor.loc[(apt2017_floor['층'] > 15) & (apt2017_floor['층'] <= 20), 'floor_group'] = "16~20층"
apt2017_floor.loc[(apt2017_floor['층'] > 20) & (apt2017_floor['층'] <= 25), 'floor_group'] = "21~25층"
apt2017_floor.loc[(apt2017_floor['층'] > 25) & (apt2017_floor['층'] <= 30), 'floor_group'] = "26~30층"
apt2017_floor.loc[apt2017_floor['층'] > 30, 'floor_group'] = "31층 이상"

# 2018 새로운 열을 추가함
apt2018_floor['floor_group'] = apt2018_floor['층']

# floor_group은 문자열로 바꿔주기
apt2018_floor['floor_group'] = apt2018_floor['floor_group'].astype(str)

apt2018_floor.loc[(apt2018_floor['층'] > 0) & (apt2018_floor['층'] <= 5), 'floor_group'] = "1~5층"
apt2018_floor.loc[(apt2018_floor['층'] > 5) & (apt2018_floor['층'] <= 10), 'floor_group'] = "6~10층"
apt2018_floor.loc[(apt2018_floor['층'] > 10) & (apt2018_floor['층'] <= 15), 'floor_group'] = "11~15층"
apt2018_floor.loc[(apt2018_floor['층'] > 15) & (apt2018_floor['층'] <= 20), 'floor_group'] = "16~20층"
apt2018_floor.loc[(apt2018_floor['층'] > 20) & (apt2018_floor['층'] <= 25), 'floor_group'] = "21~25층"
apt2018_floor.loc[(apt2018_floor['층'] > 25) & (apt2018_floor['층'] <= 30), 'floor_group'] = "26~30층"
apt2018_floor.loc[apt2018_floor['층'] > 30, 'floor_group'] = "31층 이상"

# 2019 새로운 열을 추가함
apt2019_floor['floor_group'] = apt2019_floor['층']

# floor_group은 문자열로 바꿔주기
apt2019_floor['floor_group'] = apt2019_floor['floor_group'].astype(str)

apt2019_floor.loc[(apt2019_floor['층'] > 0) & (apt2019_floor['층'] <= 5), 'floor_group'] = "1~5층"
apt2019_floor.loc[(apt2019_floor['층'] > 5) & (apt2019_floor['층'] <= 10), 'floor_group'] = "6~10층"
apt2019_floor.loc[(apt2019_floor['층'] > 10) & (apt2019_floor['층'] <= 15), 'floor_group'] = "11~15층"
apt2019_floor.loc[(apt2019_floor['층'] > 15) & (apt2019_floor['층'] <= 20), 'floor_group'] = "16~20층"
apt2019_floor.loc[(apt2019_floor['층'] > 20) & (apt2019_floor['층'] <= 25), 'floor_group'] = "21~25층"
apt2019_floor.loc[(apt2019_floor['층'] > 25) & (apt2019_floor['층'] <= 30), 'floor_group'] = "26~30층"
apt2019_floor.loc[apt2019_floor['층'] > 30, 'floor_group'] = "31층 이상"
```
